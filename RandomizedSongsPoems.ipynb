```python
import re

def format_poem(text):
    # Normalize spaces
    text = " ".join(text.split())

    # Add a line break after punctuation that typically ends a poetic line
    text = re.sub(r'([.,;:?!])\s+', r'\1\n', text)

    # OPTIONAL: break on long phrases starting with capital letters,
    # BUT only if they aren’t part of a word like "I'm" or "'Midst"
    text = re.sub(r'\s+(?=[A-Z][a-z])', r'\n', text)

    # Reduce triple newlines to double for stanza breaks
    text = re.sub(r'\n{3,}', r'\n\n', text)

    return text.strip()
```
